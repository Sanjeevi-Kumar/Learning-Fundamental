```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,10,30,20,5]
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("line plot")
plt.show()
```

```python
plt.plot(x,y,color="violet",linestyle="--",marker="o",linewidth=3,markersize=10)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(x,y,color="red")
plt.title("X vs Y")
plt.subplot(1,2,2)
plt.plot(y,x,color="blue")
plt.title("Y vs X")
plt.show()
```

**Histogram**

```python
plt.hist([1,2,3,4,5,6,6,7,2,2,5,7,7,7,7,7],bins=5)
plt.yticks(range(1, 10))
```

**Scatterplot**

```python
x=[1,2,3,4,5]
y=[1,4,6,8,10]
plt.scatter(x,y)
```

```python
import matplotlib.pyplot as plt

labels = ["A", "B", "C", "D", "E"]
data = [10, 20, 30, 40, 50]
colors = ["red", "pink", "blue", "green", "orange"]

plt.pie(data, labels=labels, colors=colors,autopct="%.1f%%")
plt.show()
```
