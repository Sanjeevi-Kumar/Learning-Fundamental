```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,10,30,20,5]
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("line plot")
plt.show()
```

```python
plt.plot(x,y,color="violet",linestyle="--",marker="o",linewidth=3,markersize=10)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(x,y,color="red")
plt.title("X vs Y")
plt.subplot(1,2,2)
plt.plot(y,x,color="blue")
plt.title("Y vs X")
plt.show()
```
